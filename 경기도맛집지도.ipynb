```python
#필요한 파이썬 모듈을 import
from google.colab import files #google colab에 파일 업로드 위한 모듈
import io #업로드된 파일을 읽기위한 모듈
import pandas as pd #데이터 분석을 위한 모듈
import folium #지도를 화면에 표시하기 위한 모듈
```

```python
#이 코드 실행 후, 아래에서 '파일 선택' 하여 '맛집현황.csv' 업로드
uploaded = files.upload()
```

```python
# 업로드된 '맛집현황.csv' 파일을 '맛집'이라는 변수로 업로드 (주의: 해당 파일은 euc-kr로 encoding되어 있어 이를 지정해줘야 정상적으로 읽을 수 있음)
맛집 = pd.read_csv(io.BytesIO(uploaded['맛집현황.csv']),encoding='euc-kr')
```

```python
#일부 column 명 변경
맛집.rename(columns={맛집.columns[7] : '위도'}, inplace=True)
맛집.rename(columns={맛집.columns[8] : '경도'}, inplace=True)
맛집.head()
```

```python
매탄고위치 = [37.262576051,127.04475557]
map = folium.Map(location=매탄고위치,zoom_start=13) #매탄고 위치의 지도를 불러옴
folium.Marker(매탄고위치,popup='매탄고').add_to(map) #매탄고 위치를 지도에 marker 로 표시
map #지도를 화면에 보여줌
```

```python
매탄고위치 = [37.262576051,127.04475557]
map = folium.Map(location=매탄고위치,zoom_start=13) #매탄고 위치의 지도를 불러옴
folium.CircleMarker(매탄고위치,radius=50,popup='매탄고반경',color='#5555ff',fill_color='#5555ff').add_to(map)  #매탄고 주변지역을 지도에 원으로 표시
map #지도를 화면에 보여줌
```

```python

맛집['거리'] = 맛집.apply(lambda x: haversine(매탄고위치, (x['위도'] ,x['경도'])), axis=1)
```

```python

#매탄고등학교로부터 가장 가까운 5개의 맛집 표시
맛집.sort_values(by='거리', ascending=True).head(5)
```
